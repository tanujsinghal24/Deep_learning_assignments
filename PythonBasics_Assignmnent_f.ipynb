Find whether a list a palindrome -> a sequence that reads the same forwards and backwards <br>
is_palindrome([1, 2, 1]) -> True <br>
is_palindrome([1, 2, 3, 2, 1] -> True <br>
is_palindrome([1, 2, 3, 1]) -> False

```python
def is_palindrome(l):
    """
    Input:
      l: Python list
    
    Output:
      boolean value indicating whether l is a palindrome or not
    
    """
    # YOUR CODE HERE
    if l == l[::-1]:
      return True
    return False
```

Replace consecutive duplicate elements of list with single element. <br>
Eg., l = [a, a, a, b, b, c, a, a, d, d, d, x, x] <br>
compress(l) -> [a, b, c, a, d, x]

```python
def compress(l):
    """
    Input:
      l: Python list
    Output:
      Returns a list with consecutive duplicate elements replaced by a single element
    """
    # YOUR CODE HERE
    new_l=l
    if len(l)>0:
      curr = l[0]
      new_l=[curr]
      for ele in l[1:]:
        if ele != curr:
          curr=ele
          new_l.append(curr)
    return new_l
```

```python
"""Testing code for compress"""

assert(compress([1, 2, 2])) == [1, 2]
assert(compress([1, 2, 2, 2, 1, 1, 3, 'x', 'x', 'x'])) == [1, 2, 1, 3, 'x']


print("Sample Tests passed", '\U0001F44D')
```

Pack consecutive duplicates of list elements into sublists. <br>
Eg., l = [1, 1, 1, 2, 2, 3, 3, 4]
pack(l) -> [[1,1,1], [2,2], [3,3], [4]]

```python
def pack(l):
    """
    Input:
      l: Python list
      
    Output:
      Returns a list with consecutive duplicate elements packed into sublists
    """
    # YOUR CODE HERE
    new_l=l
    if len(l)>0:
      curr = [l[0]]
      new_l=[]
      for ele in l[1:]:
        if ele == curr[0]:
          curr.append(ele)
        else:
          new_l.append(curr)
          curr=[ele]
      new_l.append(curr)
    return new_l
```

```python
"""Testing code for pack"""
assert(pack([1, 1, 1, 2]) == [[1, 1, 1], [2]])
assert(pack([1, 1, 1, 2, 1, 1, 3, 3, 3])) == [[1, 1, 1], [2], [1, 1], [3, 3, 3]]

print("Sample Tests passed", '\U0001F44D')
```

Given two indices, i and k, the slice is the list containing the elements between the i'th and k'th element of the original list (left limit included). Start counting the elements with 0.

Eg., l = [1, 3, 9, 8, 7];
slice(l, 1, 3) --> [3, 9]


```python
def slice(l, i, k):
    """
    Input:
      l: Python list
      i: start index of slice
      k: end index of slice (not included)
      
    Output:
      Returns a list containing the elements between i'th and k'th elements of original list l.
    """  
    # YOUR CODE HERE
    if len(l)>i:
       return l[i:k]
    return l
```

  Flatten a nested list structure.
  Transform a list, possibly holding lists as elements into a `flat' list by replacing each          
  list with its elements (recursion)
   Example:
    def my_flatten([a, [b, [c, d], e]])-->[a, b, c, d, e]


```python
def my_flatten(l):
  """
  Input: python list
  Output:
    python list with nested lists flattened
  """
  # YOUR CODE HERE
  
  flat_list = []
  for ele in l:
    if type(ele) == list:
      ele=my_flatten(ele)
      flat_list.extend(ele)
    else:
      flat_list.append(ele)
  return flat_list
```

 Rotate a list N places to the left.
 Examples:<br>
  rotate_list([a,b,c,d,e,f,g,h],3)--> [d,e,f,g,h,a,b,c]
   <br> rotate_list([a,b,c,d,e,f,g,h],-2)--> [g,h,a,b,c,d,e,f]

```python
def rotate_list(l,N):
  """
 Input:
   l: Python list
   N: number of indices by which to rotate the list
 Ouput:
   python list which is l rotated to the left by n places
 """
  # YOUR CODE HERE
  if N>len(l):
    return l
  else:
    return l[N:]+l[:N]
```

 Determine whether a given integer number is prime.
 Example:<br>
is_prime(7)-->True<br>
is_prime(10)-->False

```python
def is_prime(n):
  """
 Input:
  n: int 
 Output:
  bool indicating whether n is prime
 """
 # YOUR CODE HERE
  if n==0 or n==1:
    return False
  if n==2:
    return True
  for i in range(2,int(n/2) +1):
    if n%i==0:
      return False
  return True
```

```python
def gcd(a,b):
    """
    Input:
      a, b: ints
    Output:
      int which is the gcd of a, b
    """
    # YOUR CODE HERE
    if a > b: 
        s = a 
    else: 
        s = b 
    for i in range(1, s+1): 
        if((a % i == 0) and (b % i == 0)): 
            gcd = i 
    return gcd 
```

Write a Python program to combine two dictionary adding values for common keys.<br>
d1 = {'a': 100, 'b': 200, 'c':300}<br>
d2 = {'a': 300, 'b': 200, 'd':400}<br>Example-<br>
combine_duplicates(d1,d2)-->>({'a': 400, 'b': 400, 'd': 400, 'c': 300})

```python
def combine_duplicates(d1,d2):
 """
 Input:
  d1, d2 : Python dictionaries
 Outpu:
  Dictionary concatenating (key, value) pairs of d1, d2 and adding values of common keys
 """
 # YOUR CODE HERE
 for key in d2: 
  if key in d1: 
    d2[key]=d2[key]+d1[key]
 for key in d1:
  if key not in d2:
    d2[key]=d1[key]
 return d2
```

```python
"""Test for combine_duplicates"""
assert combine_duplicates({'a':10, 'b':20}, {'a':50, 'c':20}) == {'a':60, 'b':20, 'c':20}
print("Sample Tests passed", '\U0001F44D')
```

Write a Python program to find the highest n values in a dictionary.<br>
Example-<br>
max_3({'a':500, 'b':5874, 'c': 560,'d':400, 'e':5000, 'f': 20}-->[5874,5000,500]

```python
def max_n(d,n):
  """
  Input:
    d: python dictionary
    n: int indicating number of highest values to return from d
  Output:
    list of length n containing highest n values of d. 
    (return all values if number of values in d is less than n)
  """
  # YOUR CODE HERE
  max_list=list(d.values())
  max_list.sort(reverse=True)
  return max_list[:n]
```

```python
"""Test for max_n """
assert max_n({'a':500, 'b':5874, 'c': 560,'d':400, 'e':5000, 'f': 20},2) == [5874, 5000]
print("Sample Tests passed", '\U0001F44D')
```
