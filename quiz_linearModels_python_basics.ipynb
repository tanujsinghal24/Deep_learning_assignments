## Question 1
I have a function $y=\log x + x, x\in(1,2)$. I measured it at 3 points $x\in\{1, 1.5, 1.7\}$. Now, I thought I can fit a model $\hat{y} = w_0 + w_3 x^3$ to this data, so that I could approximate $y$ for any value of $x$. 

a. Find the optimal values of weights $w_0, w_3$ that minimize the squared error. 

b. If I test my model at $x=1.3$, what will be the squared error $E=(y-\hat{y})^2$?

You can use pen-paper to find the answers. Write your final answers below in the variables with names w_0, w_3 and E

```python
# YOUR CODE HERE
import numpy as np
x=[1,1.5,1.7]
yd=x+np.log(x)
A=[]
for e in x:
    ls=[]
    for i in range (4):
        if i in [0,3]:
            ls.append(float(e**i))
    A.append(ls)  
A=np.asarray(A)
w=np.linalg.inv(A.T.dot(A)).dot(A.T).dot(yd)
y=A.dot(w)
mse=(y-yd).T.dot((y-yd))
mse
```

```python
w_0=0.72072894
```

```python
w_3=0.32018804
```

```python
E=0.01650361321059763
```

### Sum of squares
Given  a list of floats, find  the sum of squares <br>
Eg., 
l = [1, 2, 3] <br>
sum_sqs(l) == 14

```python
def sum_sqs(l):
  """
  Input:
    l : list of ints
    y : int, sum of squares of elements of l
  """
  # YOUR CODE HERE
  y=0
  for ele in l:
    y=y+ele**2
  return y
```

### Sub Matrix
Given a matrix X (numpy array) of shape (m, n) where m>=3, n>=3. Find the first sub-matrix Y of size (3, 3)
Eg.,
$$
\begin{align}
M = \begin{bmatrix}
[11, & 12, & 13, &14], \\
[21, & 22, & 22, & 23],\\
[31, & 32, & 33, & 34],\\
[41, & 42, & 43, & 44],\\
\end{bmatrix}
\end{align}
$$
$$
\begin{align}
Y = \begin{bmatrix}
[11, & 12, & 13,], \\
[21, & 22, & 22,],\\
[31, & 32, & 33,],\\
\end{bmatrix}
\end{align}
$$

```python
def first_sub_matrix_3_3(X):
  """
  Input:
    X : numpy array of shape (m, n), m>=3, n>=3
  Output:
    Y : numpy array of shape (3, 3); first submatrix of X
  """
  # YOUR CODE HERE
  Y=X[0:3,0:3]
  return Y
```

### Windows
Given a list l, and window size m, split it into windows of size m as shown below. <br>
Eg., l = [1, 2, 3, 4, 5, 6], m = 3 <br>
get_windows(l, m) --> [ [1, 2, 3], [2, 3, 4],  [3, 4, 5], [4, 5, 6]]

```python
def get_windows(l, m):
  """
    Input:
      l : list of ints
      m : int, window_length
    Output:
      y : list of lists, each inner list is a window
  """
  # YOUR CODE HERE
  i=0
  y=[]
  while(i+m<= len(l)):
    curr=l[i:i+m]
    i+=1
    y.append(curr)
  return y
```
